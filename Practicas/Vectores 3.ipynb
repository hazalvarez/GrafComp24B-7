```python
import numpy as np # soporte para vectores
import matplotlib.pyplot as plt # graficador
```

### Quiver
La función plt.quiver de Matplotlib se usa para generar gráficos que representan campos de vectores en dos dimensiones.

### Sintaxis
plt.quiver(X, Y, U, V)

X, Y: Coordenadas de los puntos donde se dibujarán los vectores.
U, V: Componentes de los vectores en las direcciones X e Y en cada punto.

### axvline() y axhline()
Las funciones plt.axvline() y plt.axhline() de Matplotlib se emplean para trazar líneas verticales y horizontales, respectivamente, dentro de un gráfico.

```python
plt.quiver(1,2, 3,4, scale_units='xy', angles='xy', scale=1)
plt.xlim(0,4)
plt.ylim(0,8)

# guias
plt.axvline(x=1)
plt.axhline(y=2)

plt.axvline(x=1+3)
plt.axhline(y=2+4)
```

### xlim() y ylim()
Las funciones plt.xlim() y plt.ylim() de Matplotlib se utilizan para definir los límites de los ejes X e Y en un gráfico.

```python
plt.quiver([0,0],[0,0],[1,2],[4,2], scale_units='xy', angles='xy', scale=1)
plt.xlim(0,4)
plt.ylim(0,4)
```

En la gráfica anterior, se dibujan dos vectores que parten del origen (0,0) y se extienden hacia las coordenadas (1,4) y (2,2).

El plano está limitado por los valores 4 en el eje **X** y 4 en el eje **Y**.

```python
plt.quiver([1,0],[2,0],[1,2],[3,2], scale_units='xy', angles='xy', scale=1)
plt.xlim(0,4)
plt.ylim(0,6)

plt.axvline(x=1, color='gray', linestyle='--', alpha=0.3)
plt.axhline(y=2, color='gray', linestyle='--', alpha=0.3)

plt.axvline(x=2, color='gray', linestyle='--', alpha=0.3)
plt.axhline(y=2, color='gray', linestyle='--', alpha=0.3)

plt.axvline(x=2, color='gray', linestyle='--', alpha=0.3)
plt.axhline(y=5, color='gray', linestyle='--', alpha=0.3)
```

El primer vector parte de (1,2) con componentes (1,3), y el segundo comienza en (0,0) con componentes (2,2). Los parámetros `scale_units='xy'` y `angles='xy'` aseguran que las escalas y ángulos correspondan al sistema cartesiano. Los límites del gráfico se ajustan con `plt.xlim(0,4)` y `plt.ylim(0,6)`. Además, se añaden líneas punteadas y semitransparentes en **x=1**, **y=2**, **x=2** y **y=5** usando `plt.axvline()` y `plt.axhline()`.

```python
v1 = np.array([1,2])
v2 = np.array([3,4])

plt.quiver([0,0],[0,0],[v1[0],v2[0]],[v1[1],v2[1]],angles='xy', scale_units='xy', scale=1)
plt.xlim(0, max(v1[0], v2[0]))
plt.ylim(0, max(v1[1], v2[1]))

plt.axvline(x=1, alpha=0.2, linestyle='--')
plt.axhline(y=2, alpha=0.2, linestyle='--')
plt.axvline(x=3, alpha=0.2, linestyle='--')
plt.axhline(y=4, alpha=0.2, linestyle='--')
```

Este código grafica dos vectores utilizando Matplotlib. Los vectores v1 = [1, 2] y v2 = [3, 4] parten del origen (0, 0) y terminan en las coordenadas (1, 2) y (3, 4), respectivamente.

Se añaden líneas de referencia (con plt.axvline y plt.axhline) para marcar las posiciones finales de los vectores en los ejes X e Y. Estas líneas se dibujan con un estilo discontinuo y con algo de transparencia. Los límites del gráfico se ajustan automáticamente para abarcar los extremos de ambos vectores.

plt.axvline() y plt.axhline(): Dibuja líneas verticales y horizontales como referencia en las posiciones donde terminan los vectores. El parámetro alpha=0.2 añade transparencia, mientras que linestyle='--' establece el estilo de línea discontinua.

```python
def grafvecs(vecs, cols):
    plt.figure(figsize=(5,5))
    # vecs: lista de vectores
    for i in range(len(vecs)):
        vec = vecs[i]
        plt.quiver(0,0, vec[0], vec[1], color = cols[i], 
                   angles='xy', scale_units='xy', scale=1)
```

### *Explicación de la función*

La función **grafvecs(vecs, cols1)** está diseñada para graficar una lista de vectores en un plano bidimensional utilizando la función `plt.quiver()`.

La función toma dos parámetros:

- **vecs**: Una lista de vectores, donde cada vector es una lista o tupla con dos componentes, representando las coordenadas **xx** y **yy** del vector.
- **cols**: Una lista de colores, donde cada color corresponde a un vector en el gráfico.

Primero, se crea una figura de 5x5 pulgadas con `plt.figure()`, lo que define el tamaño del gráfico.

Luego, `plt.quiver()` se utiliza para dibujar un vector desde el origen (0,0) hasta las coordenadas determinadas por las componentes del vector, es decir, `vec[0]` para la componente **xx** y `vec[1]` para la componente **yy**.

- `color=cols[i]`: Asigna el color correspondiente de la lista **cols** al vector.
- `angles='xy'`: Indica que las componentes del vector están en el sistema de coordenadas cartesianas.
- `scale_units='xy'`: Mantiene las mismas unidades de escala para los ejes **X** e **Y**.
- `scale=1`: No aplica escalado adicional, por lo que las flechas tienen el tamaño exacto de las componentes del vector.

```python
v1 = np.array([1,2])
v2 = np.array([3,4])
v3 = np.array([8,4])

grafvecs([v1, v2, v3], ['blue', 'red', 'black'])
plt.xlim(0, max(v1[0], v2[0], v3[0]))
plt.ylim(0, max(v1[1], v2[1], v3[1]))
```

**Vectores:**
   - `v1 = np.array([1,2])`: El primer vector tiene las componentes **(1, 2)**.
   - `v2 = np.array([3,4])`: El segundo vector tiene las componentes **(3, 4)**.
   - `v3 = np.array([8,4])`: El tercer vector tiene las componentes **(8, 4)**.


**Configuración de límites:**
   - `plt.xlim(0, max(v1[0], v2[0], v3[0]))`: Establece el límite del eje X desde 0 hasta el valor máximo entre las primeras componentes de los vectores (**1**, **3** y **8**), por lo que el eje X va de 0 a 8.
   - `plt.ylim(0, max(v1[1], v2[1], v3[1]))`: Establece el límite del eje Y desde 0 hasta el valor máximo entre las segundas componentes de los vectores (**2**, **4**, y **4**), por lo que el eje Y va de 0 a 4.

**Gráfico:**
   La gráfica muestra tres vectores que parten del origen (0,0):
   - El vector azul va hacia la coordenada (1,2).
   - El vector rojo se dirige hacia (3,4).
   - El vector negro se extiende hasta (8,4).
